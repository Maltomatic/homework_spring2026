```python
# generate charts from /exp/*/log.csv

import matplotlib.pyplot as plt
import pandas as pd
import glob
import os

def get_log_batch(experiment_path):
    # Get the path to the CSV file
    csv_file = os.path.join(experiment_path, 'log.csv')
    
    # Check if the CSV file exists
    if os.path.isfile(csv_file):
        frame = pd.read_csv(csv_file)
        return frame
    else:
        print(f"CSV file not found in {experiment_path}")
        return None

def single_experiment(experiment_path, y):
    # Load the CSV file into a DataFrame
    csv_file = os.path.join(experiment_path, 'log.csv')
    df = pd.read_csv(csv_file)

    #compare the learning curves (average return vs. number of environment steps)
    plt.figure(figsize=(10, 6))
    plt.plot(df['Train_EnvstepsSoFar'], df[y], label=f"Steps / {y}")
    plt.xlabel('Environment Steps')
    plt.ylabel(y)
    plt.title('Learning Curve')
    plt.legend()
    plt.grid()
    plt.savefig(os.path.join(experiment_path, f'learning_curve_{y}.png'))
    print(f"Saved learning curve for {experiment_path}")
    plt.close()

def plot_experiment(logs_dict, title, y='Eval_AverageReturn'):
    plt.figure(figsize=(10, 6))
    # different color for each experiment, with legend
    for exp_name, df in logs_dict.items():
        plt.plot(df['step'], df[y], label=exp_name)

    plt.xlabel('Train_EnvstepsSoFar')
    plt.ylabel(y)
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.savefig(f'{title.replace(" ", "_").lower()}.png')
    print(f"Saved plot: {title}")
    plt.close()


# Get a list of all small experiment directories
# regex: no _lb_ in the directory name
dir_names = glob.glob('exp/*/')
```

```python
cart_dir_names = [d for d in dir_names if 'CartPole' in d]
small_batch_dirs = [d for d in cart_dir_names if 'lb' not in d]
large_batch_dirs = [d for d in cart_dir_names if 'lb' in d]

# Stack plots for small and large batch size experiments separately
small_logs = dict()
for exp_dir in small_batch_dirs:
    # drop all after 'sd' for name
    print(f"Processing {exp_dir}")
    exp_name = exp_dir.split('\\')[-2].split('_sd')[0]
    print(f"Using {exp_name}")
    small_logs[exp_name] = get_log_batch(exp_dir)
plot_experiment(small_logs, 'Small Batch', y='Eval_AverageReturn')
large_logs = dict()
for exp_dir in large_batch_dirs:
    exp_name = exp_dir.split('\\')[-2].split('_sd')[0]
    large_logs[exp_name] = get_log_batch(exp_dir)
plot_experiment(large_logs, 'Large Batch', y='Eval_AverageReturn')
```

```python
# Cheetah experiments
default_dirs = [d for d in dir_names if 'HalfCheetah' in d and 'bgs' not in d and 'blr' not in d and 'baseline' in d]
cheetah_default_logs = dict()
for exp_dir in default_dirs:
    print(f"Processing {exp_dir}")
    exp_name = exp_dir.split('\\')[-2].split('_sd')[0]
    print(f"Using {exp_name}")
    cheetah_default_logs[exp_name] = get_log_batch(exp_dir)
plot_experiment(cheetah_default_logs, 'HalfCheetah Default Param Baseline Loss', y= "Baseline Loss")
plot_experiment(cheetah_default_logs, 'HalfCheetah Default Param Eval Average Return', y= "Eval_AverageReturn")

cheetah_baseline_exp_dir = [d for d in dir_names if 'HalfCheetah' in d and 'baseline' in d]
cheetah_logs = dict()
# plot baseline loss, eval max return
for exp_dir in cheetah_baseline_exp_dir:
    print(f"Processing {exp_dir}")
    exp_name = exp_dir.split('\\')[-2].split('_sd')[0]
    print(f"Using {exp_name}")
    cheetah_logs[exp_name] = get_log_batch(exp_dir)
plot_experiment(cheetah_logs, 'HalfCheetah Baseline Loss', y= "Baseline Loss")
plot_experiment(cheetah_logs, 'HalfCheetah Eval Average Return', y= "Eval_AverageReturn")
```

```python
# LunarLander
lander_dirs = [d for d in dir_names if 'LunarLander' in d]
lander_logs = dict()
for exp_dir in lander_dirs:
    print(f"Processing {exp_dir}")
    exp_name = exp_dir.split('\\')[-2].split('_sd')[0]
    print(f"Using {exp_name}")
    lander_logs[exp_name] = get_log_batch(exp_dir)
plot_experiment(lander_logs, 'LunarLander Eval Average Return', y='Eval_AverageReturn')
```
