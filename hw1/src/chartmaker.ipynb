```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# MSE log path
MSE_rewards_csv = "./hw1_imitation/exp/mse/log.csv"
# read rewards
MSE_rewards = pd.read_csv(MSE_rewards_csv)
# plot rewards
plt.plot(MSE_rewards['step'], MSE_rewards['eval_reward'])
plt.xlabel('Step')
plt.ylabel('Eval Reward')
plt.title('MSE Rewards')
plt.show()

MSE_loss_path = "./hw1_imitation/exp/mse/mse_loss.csv"
MSE_loss = pd.read_csv(MSE_loss_path)
plt.plot(MSE_loss['step'], MSE_loss['loss'])
plt.xlabel('Step')
# show only every 100th x tick
plt.xticks(MSE_loss['step'][99::100])
plt.ylabel('MSE Loss')
plt.title('MSE Loss')
plt.show()
```

```python
# MSE log path
flow_rewards_csv = "./hw1_imitation/exp/flow/log.csv"
# read rewards
flow_rewards = pd.read_csv(flow_rewards_csv)
# plot rewards
plt.plot(flow_rewards['step'], flow_rewards['eval_reward'])
plt.xlabel('Step')
plt.ylabel('Eval Reward')
plt.title('flow Rewards')
plt.show()

flow_loss_path = "./hw1_imitation/exp/flow/flow_loss.csv"
flow_loss = pd.read_csv(flow_loss_path)
plt.plot(flow_loss['step'], flow_loss['loss'])
plt.xlabel('Step')
# show only every 100th x tick
plt.xticks(flow_loss['step'][99::100])
plt.ylabel('flow Loss')
plt.title('flow Loss')
plt.show()
```
